1. Что делает команда python -m venv venv?

Ответ: Создает виртуальное окружение с названием .venv в папке проекта.

1.1 Что делает каждая команда в списке ниже?





Ответ: pip list - показывает список библиотек и их версии, которые установлены с помощью этого пакетного менеджера.
       pip freeze > requirements.txt - эта команда экспортирует список библиотек проекта и их версий в файл requirements.txt.
       pip install -r requirements.txt - эта команда устанавливает зависимости из списка в файле requirements.txt(библиотеки).
    
2. Что делает каждая команда в списке ниже?



Ответ: conda env list - показывает список библиотек и их версии в этом пакетном менеджере(конда).
       conda create -n env_name python=3.5 - создает виртуальное окружение с помощью конды с пайтоном версии 3.5.
       conda env update -n env_name -f file.yml - обновляет текущее окружение и зависимости по файлу file.yml.
       source activate env_name - активирует уже созданное виртуальное окружение.
       source deactivate - деактивирует(выходит из него) виртуальное окружение.
       conda clean -a - удаляет кеш и временные файлы связанные с кондой.

4. Как установить необходимые пакеты внутрь виртуального окружения для conda/venv?

5. Что делают эти команды?

pip freeze > requirements.txt - эта команда экспортирует список библиотек проекта и их версий в файл requirements.txt.
conda env export > environment.yml - эта команда экспортирует окружение в файл environment.yml, чтобы потом можно было его активировать.

6. Что делают эти команды?

7. Что делают эти команды?

pip show - показывает подробную информацию о конкретной библиотеке.
conda list - показывает список библиотек и их версии в этом пакетном менеджере(конда).

8. Где по умолчанию больше пакетов venv/pip или conda? и почему дата сайнинисты используют conda?

Ответ: Conda имеет очень большой пакет библиотек из коробки, что иногда очень удобно, чтобы не тратить время на мануальную установку, а pip позволяет УСТАНОВИТЬ больше библиотек, но изначально не имеет большого количества уже установленных библиотек. ДС'ы используют conda из-за удобства также там легче настроить определенное окружение с разной версией пайтон.
