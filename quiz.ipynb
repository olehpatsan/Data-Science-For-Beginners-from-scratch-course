10.1) Напишите список линтеров которые используются для проверки кода и дайте их краткую характеристику.

Ответ: jupytext - делает копию нотбуков формата .ipynb в формате .py для применения линтер-модулей и удобного трека версий Git.
       docformatter - модуль, который форматирует docstring.

       mypy(nbQA) - проверяет Python код в .py и .ipynb(nbQA версия) на соответствие типов.
